### 1) Importação de biliotecas e dos dados 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1
df = pd.read_csv('medical_examination.csv')    
```

```python
# 2
df['overweight'] = (df['weight'] / ((df['height'] / 100) ** 2) > 25).astype(int)
```

```python
# 3

df['colesterol'] = (df['cholesterol'] > 1).astype(int)
df['gluc'] = (df['gluc'] > 1).astype(int)

df.head()
```

```python
# 4

def draw_cat_plot():
    sns.catplot(
        data = df,
        x = "cholesterol",
        hue = "cardio",
        kind = "count"
    )
    plt.show()
```

```python
# 5

df_cat = pd.melt(
    df,
    id_vars=["cardio"],
    value_vars=[
        "cholesterol",
        "gluc",
        "smoke",
        "alco",
        "active",
        "overweight"
    ]
)

df_cat.head(20)
```

```python
# 6 

df_cat = df_cat.rename(columns={"variable": "feature", "value": "value"})

df_cat = df_cat.groupby(["cardio", "feature", "value"]).size().reset_index(name="count")

df_cat.head(50)
```

```python
# 7,  8 e 9

def draw_cat_plot():
    fig = sns.catplot(
        data=df_cat,
        x="feature",
        y="count",
        hue="value",
        col="cardio",
        kind="bar"
    ).fig
    fig.savefig('catplot.png')
    return fig
```

```python
# 10
def draw_heat_map():
    # Clean the data
    df_heat = df[
        (df['ap_lo'] <= df['ap_hi']) &
        (df['height'] >= df['height'].quantile(0.025)) &
        (df['height'] <= df['height'].quantile(0.975)) &
        (df['weight'] >= df['weight'].quantile(0.025)) &
        (df['weight'] <= df['weight'].quantile(0.975))
    ]

    # Calculate the correlation matrix
    corr = df_heat.corr()

    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Set up the matplotlib figure
    fig, ax = plt.subplots(figsize=(12, 10))

    # Draw the heatmap
    sns.heatmap(
        corr,
        mask=mask,
        annot=True,
        fmt=".1f",
        center=0,
        vmax=0.3,
        vmin=-0.1,
        square=True,
        linewidths=0.5,
        cbar_kws={"shrink": 0.5}
    )

    fig.savefig('heatmap.png')
    return fig
    
```
